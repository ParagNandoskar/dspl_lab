```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./thyroid_cancer_risk_data.csv") 
df2 = df
```

```python
print(plt.figure(figsize = (16, 5), dpi = 1000))
print("after plot")
sns.boxplot(x = "Age", data = df2) 
print("after put data")
```

```python
plt.figure(figsize = (10, 4), dpi = 100) 
color_palette = sns.color_palette("dark") 
sns.set_palette(color_palette)
sns.countplot(x = "Nodule_Size", data = df2) 
```

```python
plt.figure(figsize = (16, 6), dpi = 100) 
color_palette = sns.color_palette("colorblind6")
 
sns.set_palette(color_palette) 
sns.countplot(x = "Thyroid_Cancer_Risk", data = df2) 
```

```python
grp = dict(df2.groupby('Thyroid_Cancer_Risk').groups) 
m = {}
for key, val in grp.items(): 
    if key in m:
        m[key] += len(val) 
    else:
        m[key] = len(val) 
        plt.title("Distribution of Thyroid_Cancer_Risk") 
        plt.pie(m.values(), labels = m.keys())
```

```python
plt.figure(figsize = (10, 4), dpi = 100)
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "Gender", data = df2) 
```

```python
plt.figure(figsize = (10, 14), dpi = 100) 
color_palette = sns.color_palette("plasma") 
sns.set_palette(color_palette)
 
sns.barplot(y = "Gender", x = "Country", data = df2) 
```
